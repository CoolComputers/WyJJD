```python
import csv
import numpy as np

data_folder = '../new-data/'
dci_data = data_folder+'DCI/combined/combined-data.csv'
dfs_afcars_data = data_folder+'DFS/DFS_12-16-AFCARS-CLEAN.csv'
dfs_county_data = data_folder+'DFS/DFS_12-16-Placements-ByCounty-CLEAN.csv'
dfs_plc_data = data_folder+'DFS/DFS_12-16-Placements-ByPLC-CLEAN.csv'
school_discipline_data=data_folder+'school-discipline/SchoolDiscipline_2007-16_Combined-CSV.csv'

def createDataObject(file):
    with open(school_discipline_data) as csvfile:
        school_discipline_reader = csv.reader(csvfile, delimiter=',', quotechar='|')
        row_number = 0
        data_object = []
        columns = []
        for row in school_discipline_reader:
            #print(', '.join(row))
            if row_number==0:
                columns = row
            else:
                tmp = {}
                for col_label,row_value in zip(columns,row):
                    tmp[col_label]=row_value
                data_object.append(tmp)
            row_number+=1
        return data_object
            
#print str(createDataObject(school_discipline_data))
print str(createDataObject(school_discipline_data)[:,1])
```
