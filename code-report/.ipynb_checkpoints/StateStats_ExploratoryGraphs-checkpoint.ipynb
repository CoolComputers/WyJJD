```python
#import plotly.plotly as py
#import plotly.graph_objs as go

#display(HTML(school_county_rates[['Total','Population','Rate']].loc[idx[:,YEAR:YEAR], :].to_html()))

#gather years
years = [x for x in range(int(BEGIN_YEAR),int(END_YEAR))]

#gather sums arrays for each year
sum_total = []
for year in years:
    sum_total.append(school_county_rates[['Rate']].loc[idx[:,year:year], :].values.flatten())
#print(sum_total)
#gather counties array
counties = school_county_rates.loc[idx[:,year:year], :].index.values.tolist()
counties = [x[0] for x in counties]
#print(counties)

#create bars for each year
data = []
marker_colors = ['#6699ff','#00ff99','#669900','#ffcc00','#ff6600','#ff6600']
cntr = 0
for year in years:
    year_avg = np.mean(sum_total[cntr])
    data.append(go.Bar(
        x = counties,
        y = sum_total[cntr],
        text=str(year),
        textposition = 'auto',
        name=year,
        marker=dict(
            color=marker_colors[cntr],
            line=dict(
                color='rgb(8,48,107)',
                width=1.5),
        ),
        opacity=0.6
        )
    )
    data.append(go.Scatter(
            x = counties,
            y = [year_avg for x in range(len(counties))],
            name='{} Average'.format(year),
            marker=dict(
            color=marker_colors[cntr],
            line=dict(
                color='rgb(8,48,107)',
                width=1.5),
        )
        )
    )
    
    cntr+=1
    
    
    
    

pl.offline.iplot(data, filename='county school rates over 5 years')
```

```python
objs = []
data = []
for county in counties:
    pct_change_vals=school_county_rates[['Rate']].loc[idx[county,'2012':'2017'], :].pct_change().values.flatten().tolist()
    objs.append({'county':county,'pct_change':pct_change_vals})
    cnt=0
    pos_years = []
    pos_data = []
    neg_years = []
    neg_data = []
    for year in years:
        if pct_change_vals[cnt]<0:
            neg_years.append(year)
            neg_data.append(pct_change_vals[cnt])
        else:
            pos_years.append(year)
            pos_data.append(pct_change_vals[cnt])
        cnt+=1
#print(objs)

cntr=0
for year in years:
    year_avg = np.mean([x['pct_change'][cntr] for x in objs])
    data.append(go.Bar(
        x = counties,
        y = [x['pct_change'][cntr] for x in objs],
        #text=,
        textposition = 'auto',
        name='{} Percent Change'.format(year),
        marker=dict(
            color=marker_colors[cntr],
            line=dict(
                color='rgb(8,48,107)',
                width=1.5),
        ),
        opacity=0.6
        )
    )
    
    data.append(go.Scatter(
            x = counties,
            y = [year_avg for x in range(len(counties))],
            name='{} Average'.format(year),
            marker=dict(
            color=marker_colors[cntr],
            line=dict(
                color='rgb(8,48,107)',
                width=1.5),
        )
        )
    )
    
    cntr+=1


pl.offline.iplot(data, filename='county school pct change in rates over 5 years')

#print(school_county_rates.head(3))
# for year in years:
#     data.append(go.Bar(
#         x = counties,
#         y = sum_total[cntr],
#         text=str(year),
#         textposition = 'auto',
#         name=year,
#         marker=dict(
#             color=marker_colors[cntr],
#             line=dict(
#                 color='rgb(8,48,107)',
#                 width=1.5),
#         ),
#         opacity=0.6
#         )
#     )
#     cntr+=1
    


# pl.offline.iplot(data, filename='county school rates over 5 years')
```
