```python
#now combine raw files into two workable .csvs, age_group1.csv and age_group2.csv
from os import listdir
from os.path import isfile, join
agepath1 = './data/age_group1/'
onlyfiles = [f for f in listdir(agepath1) if isfile(join(agepath1, f))]
print(onlyfiles)

outf = open(join('./data/','age_group1.csv'),'w+')
outf.write('Year,Age Range,County,White,Black,American Indian,Asian,Total\n')
for file in onlyfiles:
    with open(join(agepath1,file),'r') as inf:
        state=0
        for row in inf:
            print(row)
            print('row length={}'.format(len(row)))
            #determine state
            if "Year" in row:
                state=1
                year = row.split('=')[1].strip().strip('"')
            elif "Albany" in row:
                state=2
            elif "Suggested" in row:
                state=3
                
            #statework
            if state==1:
                print('Found year:{}'.format(year))
            elif state==2:
                if len(row)!=1:
                    outf.write('{},{},{}'.format(year,'11-17',row))
outf.close()

agepath2 = './data/age_group2/'
onlyfiles = [f for f in listdir(agepath2) if isfile(join(agepath2, f))]
print(onlyfiles)

outf = open(join('./data/','age_group2.csv'),'w+')
outf.write('Year,Age Range,County,White,Black,American Indian,Asian,Total\n')
for file in onlyfiles:
    with open(join(agepath2,file),'r') as inf:
        state=0
        for row in inf:
            print(row)
            print('row length={}'.format(len(row)))
            #determine state
            if "Year" in row:
                state=1
                year = row.split('=')[1].strip().strip('"')
            elif "Albany" in row:
                state=2
            elif "Suggested" in row:
                state=3
                
            #statework
            if state==1:
                print('Found year:{}'.format(year))
            elif state==2:
                if len(row)!=1:
                    outf.write('{},{},{}'.format(year,'18-20',row))
outf.close()

#then hand combined the two files into data/all_groups.csv
```
