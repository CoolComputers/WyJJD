```python
#download the data
import requests,time

BASE = 'https://www.ojjdp.gov/ojstatbb/ezapop/asp/profile_display.asp'
clean_folder = './cleaning/'

counties = [i for i in range(1,47,2)]
years = [i for i in range(11,128)]
print(counties)
print(years)

#'$BASE?selState=56&selCounty=1&selLowerYear=1&selUpperYear=1&selLowerAge=1&selUpperAge=1&row_var=v05&col_var=v03&display_type=counts&export_file=yes&printer_friendly=&v0127=v0127'
for county in counties:
    for year in years:
        try:
            req_string = '{}?selState=56&selCounty={}&selLowerYear=1&selUpperYear=1&selLowerAge=1&selUpperAge=1&row_var=v05&col_var=v03&display_type=counts&export_file=yes&printer_friendly=&v0{}=v0{}'.format(BASE,county,year,year)
            print('requesting for:{}-{}------{}'.format(county,year,req_string))
            r = requests.get(req_string)
            open('{}dirty1-{}-{}.csv'.format(clean_folder,county,year), 'wb').write(r.content)
            time.sleep(5)
        except Exception as e:
            print(str(e))
```
