```python
def getCounties():
    print('getting county links')
    county_page = 'http://eadiv.state.wy.us/demog_data/County_Profile.html'

    req = requests.get(county_page)
    soup = bs(req.content,'html.parser')
    county_tables = soup.find_all('table')
    print(len(county_tables))

    links = county_tables[4].find_all('a')
    #print(set(links))
    uniqs=set(links)
    county_links = []
    for uniq in uniqs:
        county_links.append(uniq['href'])
        #print(uniq)
        #print(uniq['href'])
    splits = []
    for county in county_links:
        #print(str(county.split('2017')))
        splits.append(county.split('2017'))
    #print(str(splits))
    return splits
```

```python
from bs4 import BeautifulSoup as bs
import requests, pandas, time
import csv

#soup = BeautifulSoup(html_doc, 'html.parser')

#http://eadiv.state.wy.us/Wy_facts/Big_Horn2015.htm
counties = getCounties()
# print(counties)
years = ['2012','2013','2014','2015','2016']

for year in years:
    for county in counties:
        print county
        #req_page = 'http://eadiv.state.wy.us/Wy_facts/Big_Horn'+year+'.htm'
        req_page = county[0]+year+county[1]
        print('requesting page:'+req_page)
        time.sleep(2)
        req = requests.get(req_page)
        if req.status_code is not 200:
            print('badreq-'+req.status_code)
            continue
        if req.status_code is 200:
            #print(req.content)
            soup = bs(req.content,'html.parser')
            data = []
            table = soup.find('table')
            if table is None: 
                continue
            #print(str(table))
            heads = table.find_all('td')
            print(len(heads))
            rows = table.find_all('tr')
            for row in rows:
                cols = row.find_all('td')
                cols = [ele.text.strip().encode('utf-8') for ele in cols]
                data.append([ele for ele in cols if ele]) 
            
            i = 8  # skip over first 8 header rows
            while i < len(data)-15:  # the 14 last lines are trash
                row = data[i]
                # print row
                if len(row) > 1:
                    # new header - create new csv file
                    if row[0].isupper():                  
                        title = county[0]
                        title = str(title.split('/')[-1])
                        title = str(title.split('.')[0])+'_'+year+'_'+"_".join(row[0].split())+'.csv'           
                        title = "_".join(title.split("/"))
                    
                with open(title, 'a+') as f:
                    writer = csv.writer(f)
                    writer.writerow(row)
                i += 1
                            
```
