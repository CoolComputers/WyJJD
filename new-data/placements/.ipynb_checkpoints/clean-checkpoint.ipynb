```python
#running all cells should result in one final file: 'combined.csv'
import re
import pandas as pd

input_file_template = 'correct/probation-jan1_2012-may31_2018-district'#<0-9>.csv

for i in range(1,10):
    #print(i)
    in_file = "{}{}.csv".format(input_file_template,i)
    print(in_file)
    inf = open(in_file,'r')
    line_number = 0
    clean_line_list = []
    state = 0 #beginning state, read/skip headers
    judicial_district = '0'
    for line in inf:
        line_number += 1
        #print("{}:{}".format(line_number,line))
        #determine state
        if "Report Period" in line and state ==0:
            state=1 #start saving these lines
        elif "Legend" in line and state==1:
            state=2 # reached the legend, read till last line
        elif "Judicial District Equals" in line and state==2:
            state=3 #save judicial district number
            
        #state work
        if state == 1:
            clean_line_list.append(line)
        elif state == 2:
            x=1#NOP
        elif state == 3:
            judicial_district = line.split(',')[0].split(' ')[-1]
    print('read file for district:{}'.format(judicial_district))
    #for clean_line,clean_line_num in zip(clean_line_list,[x for x in range(0,len(clean_line_list))]):
        #print("{}:{}".format(clean_line_num,clean_line.strip()))
    inf.close()
    outf1 = open('cleaning/clean1-district{}.csv'.format(i),'w+')
    for line in clean_line_list:
        #e.sub(r'\D', "", phone)
        clean_line = re.sub(r'[0-9]+\.[0-9]+\%','',line)#remove percentages
        clean_line = re.sub(r'[,]+',',',clean_line)#remove empty cells
        outf1.write(clean_line)
    outf1.close()
```

```python
input_file_template = 'cleaning/clean1-district'#<0-9>.csv
output_file_template = 'cleaning/clean2-district'#<0-9>.csv
all_combined = {}

for i in range(1,10):
    print('*******************************8')
    in_file = "{}{}.csv".format(input_file_template,i)
    out_file = "{}{}.csv".format(output_file_template,i)
    print(in_file)
    print(out_file)
    inf = open(in_file,'r')
    
    state = 0
    all_periods = []
    current_periods = []
    combined = {}
    
    for line in inf:
        if "Report Period" in line and state==0:
            state = 1 #save reporting periods, start new object array
        elif "Date data are based" in line and state==2:
            state = 3 #finish up object array
            
        #state work
        if state==1:
            current_periods = line.strip().split(',')[1:-1]
            
            print('current heads: {}'.format(str(current_periods)))
            for period in current_periods:
                all_periods.append(period)
                combined[period] = {}
            state=2 #done parsing headers, move on
        elif state==2:
            x=1#NOP
            for key in range(len(current_periods)):
                #print(key)
                combined[current_periods[key]][line.split(',')[0]] = line.split(',')[key+1]
                combined[current_periods[key]]['district'] = str(i)
        elif state==3:
            #done reading this chunk, ready for next chunk
            state=0
    #print(combined)        
    inf.close()
    #for key in combined.keys():
        #all_combined[key] = 
    
    outf = open(out_file,'w+')
    print(all_periods)
    all_fields = combined[all_periods[0]].keys()
    print(all_fields)
    all_districts = [str(x) for x in range(1,10)]
    print(all_districts)
    #print(combined)
    outf.write('period,{}\n'.format(','.join(all_fields)))
    for period in all_periods:
        #for district in all_districts:
            data_string = ','.join([combined[period][x] for x in all_fields])
            #print(data_string)
            outf.write('{},{}\n'.format(period,data_string))
    outf.close()
```
